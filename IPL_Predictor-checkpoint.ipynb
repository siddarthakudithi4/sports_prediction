```python
second_innings['current_Score'] = second_innings.groupby('match_id')['total_runs_y'].cumsum()
#Ensure 'total_runs_y' is numeric using to_numeric with errors='coerce' to handle potential non-numeric values:
second_innings['total_runs_y'] = pd.to_numeric(second_innings['total_runs_y'], errors='coerce')
```

### Current Run Rate

```python
second_innings['crr'] = round((second_innings['current_Score']*6)/(120 - second_innings['balls_left']),2)
second_innings
```

### Required Run Rate

```python
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
second_innings
```

### finding winner

```python
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
second_innings['results'] = second_innings.apply(winner,axis=1)
second_innings
```

### Only Taking required columns

```python
required_data = second_innings[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','total_runs_x','crr','rrr','results']]
required_data
```

### Sampling Data

```python
final_data = required_data.sample(required_data.shape[0])
```

### Dropping NUll Values

```python
final_data.dropna(inplace=True)
```

### Removing rows where balls_left==0 to remove -inf and inf from rrr column

```python
final_data = final_data[final_data['balls_left']!=0]
```

### Storing data in new csv

```python
final_data.to_csv("final_data.csv", index=False)
```

### Building Model

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x = final_data.iloc[:, :-1]
y = final_data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.25,random_state=100)
x_train
```

```python
x_test
```

```python
# from sklearn.compose import ColumnTransformer
# from sklearn.preprocessing import OneHotEncoder
# trf = ColumnTransformer([
#     ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
# ],
# remainder='passthrough')



from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Updated OneHotEncoder to use 'sparse_output' instead of 'sparse'
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
],
remainder='passthrough')

# Now you can proceed with fitting and transforming the data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[0]
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
